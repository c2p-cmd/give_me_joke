```python
import requests
import json
import time

jokes_json = []
for i in range(1, 26):
    body = requests.get(f"https://icanhazdadjoke.com/search?page={i}&limit=30", headers={
        "Accept" : "application/json"
    }).content
    json_body = json.loads(body.decode('utf-8'))
    print(len(jokes_json))
    jokes_json += json_body['results']
    print(len(jokes_json))
    print(json_body['total_pages'])
    print(json_body['next_page'])
    print("")
    time.sleep(1)
```

```python
print(len(jokes_json))

def mapper(element):
    return {
        'id' : element['id'],
        'category' : 'dad',
        'joke' : element['joke']
    }

jokes_json = list(map(mapper, jokes_json))

print(len(jokes_json))
jokes_json[0:2]
```

```python
with open("../outputs/non_programming_jokes_combined.json", "r") as jokes_file:
    current_content = json.load(jokes_file)

print(len(current_content), len(jokes_json))
current_content += jokes_json
print(len(current_content))

with open("../outputs/non_programming_jokes_all.json", "w") as all_jokes_file:
    lines = json.dumps(current_content, indent=4, sort_keys=True)
    all_jokes_file.writelines(lines)
```
